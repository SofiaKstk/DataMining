### Creating Wordclouds for Positive, Negative and Neutral Words

```python
#Now we open our vocabulary and convert it into tuples
f = open(Vocab,'r')
c = csv.reader(f,delimiter='\t') #Use the delimiter from the file , if a single space, use a single space, etc.
#for row in c:
#   print (row) Prints ['abandon', '-2'] one each loop
lexica = []
for row in c:
    lexica.append(tuple(row)) #tuple(map(str, row) same with just tuple(row)??
    #print(tuple(map(str,row)))
    #print(tuple(row))
lexica = tuple(lexica) #Instead of a list it's a tuple now.
#print("Lexica is",lexica)
```

```python
#Create vocabularies for positive, negative and neutral words
positive_words = []
negative_words = []
neutral_words = []
for word in words:
    for l in range(len(lexica)):
        if word == lexica[l][0]:
            if lexica[l][1] > str(0):              #L element, get sentiment value.
                positive_words.append(word)
            else:
                negative_words.append(word)
#            else:
#             neutral_words.append(word)
sentiment_words = positive_words + negative_words
neutral_words=[word for word in words if word not in sentiment_words]
```

### Apply KNN for K=1 to K=15 and return success rates with tfidf (or bag of words)

```python
from sklearn.neighbors import KNeighborsClassifier

Location1 = r"/home/me/Semester6/Data Mining/Project/twitter_data/train2017.tsv"
Location2 = r"/home/me/Semester6/Data Mining/Project/twitter_data/SemEval2017_task4_subtaskA_test_english_gold.txt"
f1 = open(Location1, encoding="utf8")
f2 = open(Location2, encoding="utf8")
tweet_category_train=np.array([line.split("\t")[2] for line in f1.readlines()])
tweet_category_test=np.array([line.split("\t")[1].strip('\n') for line in f2.readlines()]) #there are stray '\n's
    
def apply_knn(vector_method,k,tweet_category_train,tweet_category_test):
    if vector_method=='bag':  
        bow_vectorizer = CountVectorizer(max_df=1.0, min_df=1, max_features=1000, stop_words='english') #Bag options
        bow_xtrain = bow_vectorizer.fit_transform(tweet_list_filtered)  #object_train
        bow_xtest = bow_vectorizer.fit_transform(tweet_list_filtered_test)  #object_test
    elif vector_method=='tfidf': 
        tfidf_vectorizer = TfidfVectorizer(max_df=1.0, min_df=1, max_features=1000, stop_words='english') #tfidf options
        tfidf_train = tfidf_vectorizer.fit_transform(tweet_list_filtered)  #object for train
        tfidf_test = tfidf_vectorizer.fit_transform(tweet_list_filtered_test)  #object for test
    else:
        print('Wrong vector method!')
        return
        
    if vector_method=='bag':
        tweet_data_train = bow_xtrain.toarray()
        tweet_data_test=bow_xtest.toarray()
    elif vector_method=='tfidf': 
        tweet_data_train = tfidf_train.toarray()
        tweet_data_test  = tfidf_test.toarray()

    knn = KNeighborsClassifier(n_neighbors=k) #options
    knn.fit(tweet_data_train , tweet_category_train)  #train. Also called knn.fit(X,Y)

    tweet_category_test_results=knn.predict(tweet_data_test)
    print('Result is ', metrics.accuracy_score(tweet_category_test, tweet_category_test_results))
    
    path=r"predictions/"+"tfidf_KNN="+str(k)
    with open(path, 'wb') as fp:
        pickle.dump(tweet_category_test_results, fp)
    
    return metrics.accuracy_score(tweet_category_test, tweet_category_test_results)

a=[(k,apply_knn('tfidf',k,tweet_category_train,tweet_category_test)) for k in range(1,15)]
a
```

### Run Word2Vec 200 features

### Calculating Average of vectors of words in each tweet in order to make vector for the tweet

```python
from gensim.models import Word2Vec

tweet_list_filtered_train=tweet_list_filtered

tweet_list_filtered_train_tokenized=[tweet.split() for tweet in tweet_list_filtered_train]
tweet_list_filtered_test_tokenized =[tweet.split() for tweet in tweet_list_filtered_test]
tweets_combined_tokenized = tweet_list_filtered_train_tokenized+tweet_list_filtered_test_tokenized

model_w2v = Word2Vec(
            tweets_combined_tokenized,
            size=200, # desired no. of features/independent variables
            window=5, # context window size
            min_count=1,
            sg = 1, # 1 for skip-gram model
            hs = 0,
            negative = 10, # for negative sampling
            workers= 2, # no.of cores
            seed = 34)
model_w2v.train(tweets_combined_tokenized, total_examples= len(tweets_combined_tokenized), epochs=20)

tweets2vec_train=[]  #Tweets to vector.
for tokenized_tweet in tweet_list_filtered_train_tokenized:
    tweet2vecss=[model_w2v[word] for word in tokenized_tweet if word in model_w2v]
    if (tweet2vecss==[]):    
        #tweet2vec = 2*np.random.random((200,))-1
        tweet2vec = np.zeros((200,))
    else:    
        tweet2vec=sum(tweet2vecss)/float(len(tweet2vecss))
    tweets2vec_train.append(tweet2vec)
    
tweets2vec_test=[]  #Tweets to vector.
for tokenized_tweet in tweet_list_filtered_test_tokenized:
    tweet2vecss=[model_w2v[word] for word in tokenized_tweet if word in model_w2v]
    if (tweet2vecss==[]):    
        #tweet2vec = 2*np.random.random((200,))-1
        tweet2vec = np.zeros((200,))
    else:    
        tweet2vec=sum(tweet2vecss)/float(len(tweet2vecss))
    tweets2vec_test.append(tweet2vec)
```

### Running KNN for word2vec for k=500
